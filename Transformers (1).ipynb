```python
from  transformers import pipeline
```

```python
client=pipeline('text-generation')
text="the weather is"
client(text,
       max_length=50,
       temperature=0.8,
       top_p=1.0)
```

```python
qans=pipeline('question-answering')
context="""
H D Devegowda is the founder of Janata Dal Party.Presently it
is taking care  by  H D Kumarswamy"""
question="who is the founder of Janata Dal Party?"
result=qans(question=question, context=context)
print(result['answer'])
```

```python
ob2=pipeline('zero-shot-classification')
text="the government announces new policies related to income tax "
labels=['politics','education','finance','technology']
res=ob2(text,labels)
print(res)
```

```python
ob1=pipeline('sentiment-analysis')
ob1("sunil is a bad student")
```

```python


translator = pipeline("translation_en_to_fr")

result = translator("Suresh is feeling sleepy")

print(result)
```

```python
from transformers import pipeline

summarizer = pipeline("summarization")

text = """Born and raised in a Muslim family in Rameswaram, Tamil Nadu, Kalam studied physics and aerospace engineering...
He was buried with full state honours."""

summary = summarizer(text, max_length=50)

print(summary)
```

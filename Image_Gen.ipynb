```python
import os
os.environ["MISTRAL_API_KEY"]="4l6LOQliQQQdVIZfIPnRXjxjsnZAlwJ6"
api_key=os.environ["MISTRAL_API_KEY"]
from mistralai import Mistral
ob1=Mistral(api_key)
res=ob1.beta.agents.create(
 model="mistral-medium-2505",
 name="Image Generation Agent",
 description="Agent used to generate images",
 instructions="use image generation tools to generate images",
 tools=[{"type":"image_generation"}],
 completion_args={"temperature":0.8,
 "top_p":1.0}
)
```

```python
response=ob1.beta.conversations.start(
 agent_id=res.id,
 inputs="Rahul Gandhi winning victory"
)
```

```python
from mistralai.models import ToolFileChunk
count = 0
for chunk in response.outputs[-1].content:
  if isinstance(chunk, ToolFileChunk):
    file = ob1.files.download(file_id=chunk.file_id).read()
    open("image" + str(count) + ".png", "wb").write(file)
    count += 1
```

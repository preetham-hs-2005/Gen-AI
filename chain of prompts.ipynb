```python
client=Mistral(API_KEY)
res=client.chat.complete(
    model="mistral-medium",
    messages=[{"role":"user","content":"list of kannada movies done by puneeth rajkumar"}],
    temperature=0.8,
    top_p=1.0,
    max_tokens=1000

)
```

```python
res1=(res.choices[0].message.content)
print(res1)
```
